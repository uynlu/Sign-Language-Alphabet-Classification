```python
def visualize_label_counts(path, folders):
    label_counts = []

    for folder in folders:
        folder_path = os.path.join(path, folder)
        if os.path.exists(folder_path):
            files = os.listdir(folder_path)
            label_counts.append(len(files))
        else:
            label_counts.append(0)

    plt.figure()
    plt.bar(folders, label_counts, color = 'skyblue')
    plt.xlabel('Labels')
    plt.ylabel('Label_counts')
    plt.title('Label_counts')
    plt.xticks()
    plt.tight_layout()
    plt.show()
```

```python
def delete_unnecessary_folders(path, dirs_to_remove):
    for dir_name in dirs_to_remove:
        dir_path = os.path.join(path, dir_name)
        if os.path.exists(dir_path):
            shutil.rmtree(dir_path)
            print(f'Đã xóa thư mục {dir_path}')
        else:
            print(f'Thư mục {dir_path} không tồn tại')
```

### Extract features

```python
map_dict = {
    'A': 1,
    'B': 2,
    'C': 3,
    'D': 4,
    'E': 5,
    'F': 6,
    'G': 7,
    'H': 8,
    'I': 9,
    'J': 10,
    'K': 11,
    'L': 12,
    'M': 13,
    'N': 14,
    'O': 15,
    'P': 16,
    'Q': 17,
    'R': 18,
    'S': 19,
    'T': 20,
    'U': 21,
    'V': 22,
    'W': 23,
    'X': 24,
    'Y': 25,
    'Z': 26,
    '0': -1,
    '1': -1,
    '2': -1,
    '3': -1,
    '4': -1,
    '5': -1,
    '6': -1,
    '7': -1,
    '8': -1,
    '9': -1
}
```

```python
def extract_features(src, map_dict):
    mp_hands = mp.solutions.hands
    mp_drawing = mp.solutions.drawing_utils
    mp_drawing_styles = mp.solutions.drawing_styles

    hands = mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7)

    data = []
    labels = []

    for dir_ in os.listdir(src):
        if dir_ == 'Blank' or dir_ == '10':
            continue;
        for img_path in os.listdir(os.path.join(src, dir_)):
            data_aux = []

            x_ = []
            y_ = []

            img = cv2.imread(os.path.join(src, dir_, img_path))
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            results = hands.process(img_rgb)
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    handedness = hand_landmarks.landmark[mp_hands.HandLandmark.WRIST].x < 0.5
                    if handedness == False:
                        img_rgb = cv2.flip(img_rgb, 1)

            results = hands.process(img_rgb)
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    for i in range(len(hand_landmarks.landmark)):
                        x = hand_landmarks.landmark[i].x
                        y = hand_landmarks.landmark[i].y
                        x_.append(x)
                        y_.append(y)

                for i in range(len(hand_landmarks.landmark)):
                    x = hand_landmarks.landmark[i].x
                    y = hand_landmarks.landmark[i].y
                    data_aux.append(x - min(x_))
                    data_aux.append(y - min(y_))

                data.append(data_aux)

                label = map_dict[dir_]
                labels.append(label)

    return data, labels
```

```python
X_grass_alphabet, y_grass_alphabet = extract_features('/content/data/asl-alphabet/asl_alphabet_train/asl_alphabet_train', map_dict)
```

```python
X_test_synthetic_alphabet, y_test_synthetic_alphabet = extract_features('/content/data/synthetic-asl-alphabet/Test_Alphabet', map_dict)
```

```python
X_train_numbers, y_train_numbers = extract_features('/content/data/synthetic-asl-numbers/Train_Nums', map_dict)
```
